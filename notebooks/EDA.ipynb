```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import random
import cv2

base_dir = Path("../data/train")

classes = ["cat", "dog"]

# Conta immagini per classe
counts = {cls: len(os.listdir(base_dir / cls)) for cls in classes}
print("Numero immagini per classe:", counts)

# Istogramma distribuzione
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Distribuzione immagini per classe")
plt.show()

# Mostra alcune immagini random
for cls in classes:
    sample = random.choice(os.listdir(base_dir / cls))
    img = cv2.imread(str(base_dir / cls / sample))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(cls)
    plt.axis("off")
    plt.show()
```
